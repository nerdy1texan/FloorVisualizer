#### Drawing Different Shapes On Images

```python
### Blank White Image of Size (512x512)
img = np.ones((512,512,3),np.uint8) * 255
### Draw a line
## img,point1,point2,color,width
cv2.line(img,(50,300),(300,20),(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
img
```

```python
### Blank White Image of Size (512x512)
img = np.ones((512,512,3),np.uint8) * 255
### Draw a rectangle
## img,point1,point2,color,width
cv2.rectangle(img,(50,300),(300,20),(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
### Blank White Image of Size (512x512)
img = np.ones((512,512,3),np.uint8) * 255
### Draw a rectangle
## img,point1,point2,color,width
cv2.rectangle(img,(50,300),(300,20),(255,0,0),-1)
plt.imshow(img)
plt.show()
```

```python
### Blank White Image of Size (512x512)
img = np.ones((512,512,3),np.uint8) * 255
### Draw a Circle
## img,center,radius,color,width
cv2.circle(img,(250,250),200,(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
### Blank White Image of Size (512x512)
img = np.ones((512,512,3),np.uint8) * 255
### Draw a ellipse
## img, center, axes, angle, startAngle, endAngle, color, thickness
cv2.ellipse(img,(250,250),(200,50),0,0,360,(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
cv2.polylines?
```

```python
### Blank White Image of Size (512x512)
img = np.ones((512,512,3),np.uint8) * 255
### Draw a ellipse
## img, pts, isClosed, color[, thickness[
points = np.array([[10,200],[100,100],[450,300]])
cv2.polylines(img,[points],True,(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
cv2.putText?
```

```python
### Blank White Image of Size (512x512)
img = np.ones((1024,1024,3),np.uint8) * 255
### Write a text
## img, text, org, fontFace, fontScale, color[, thickness[,
cv2.putText(img,"Machine Learning",(20,300),cv2.FONT_HERSHEY_SCRIPT_COMPLEX,3.5,(255,0,0),2)
plt.imshow(img)
plt.show()
```
